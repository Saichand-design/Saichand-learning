```python
#Data Preprocessing
```

```python
processing_df = df_temp.copy()

dummy_Var1 = pd.get_dummies(processing_df['Education'], drop_first=True, dtype=int)
processing_df = pd.concat([processing_df, dummy_Var1], axis=1)
processing_df = processing_df.rename(columns={2: "Edu2", 3 : "Edu3"})

dummy_var2 = pd.get_dummies(processing_df['Family'], drop_first=True, dtype=int)
processing_df = pd.concat([processing_df, dummy_var2], axis=1)
processing_df = processing_df.rename(columns={2: "Fam2", 3 : "Fam3", 4 : "Fam4"})

# Convert binary categorical columns back to int type
processing_df['Securities Account'] = processing_df['Securities Account'].astype(int)
processing_df['CD Account'] = processing_df['CD Account'].astype(int)
processing_df['CreditCard'] = processing_df['CreditCard'].astype(int)

# Drop the redundant 'online' column and the original 'Family' and 'Education' columns
df = processing_df.drop(columns=['Family', 'Education', 'online'], axis = 1)
df.head()
```

```python
from scipy.stats import zscore
Vars = ['Mortgage','Age','Experience','Income','CCAvg']
X = df[Vars]
df[Vars] = X.apply(zscore)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_data = df.drop(['Personal Loan'], axis=1)
y_data = df['Personal Loan']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data,y_data,test_size=0.4,random_state=1)
```

```python
df_ytrain = y_train.value_counts().to_frame()
df_ytest = y_test.value_counts().to_frame()
print("train Data : Ratio of people who have availed Loan", df_ytrain.iloc[1]/(df_ytrain.iloc[0]+df_ytrain.iloc[1]))
print("test Data : Ratio of people who have availed Loan", df_ytest.iloc[1]/(df_ytest.iloc[0]+df_ytest.iloc[1]))
```

```python
from imblearn import over_sampling
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

oversample = SMOTE()
X_train, y_train = oversample.fit_resample(X_train, y_train)

df_ytrain = y_train.value_counts().to_frame()
print("train Data : Ratio of people who have availed loan", df_ytrain.iloc[1]/(df_ytrain.iloc[0]+df_ytrain.iloc[1]))
```

```python
# Model Evalution
```

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV,StratifiedKFold

param_grid = {
    'penalty'  : ['l1','l2'],
     'C' : np.logspace(-2,4,10),
     'solver' : ['liblinear']
}

CV = StratifiedKFold(n_splits=10,

                    shuffle = True,
                    random_state = 2)

```

```python
model = GridSearchCV(LogisticRegression(), param_grid, CV=CV, scoring = 'recall')
model.fit(X_train, y_train)
print("best Parameter: " + str(model.best_params_))

y_predict = model.predict(X_test)
# The method name 'Score' should be 'score'
model_score = model.Score(X_test, y_test)
print("ROC AUC : " + str(model_score))


cm = metrics.confusion_matrix(y_test, y_predict,labels = [1,0])
print("confusion Matrix:")
print(cm)

print("final Metrics:")
```
